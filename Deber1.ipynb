## Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%. 

```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np

# Función para generar datos de ventas aleatoriamente
def generar_ventas():
    productos = ['Producto A', 'Producto B', 'Producto C']  # Lista de productos
    datos = []

    for dia in range(1, 6):  # Últimos 5 días
        for producto in productos:
            unidades = np.random.randint(1, 21)  # Número de unidades entre 1 y 20
            precio_unitario = np.random.uniform(5, 50)  # Precio unitario entre 5 y 50
            total_venta = unidades * precio_unitario
            descuento = total_venta * 0.05
            total_con_descuento = total_venta - descuento
            datos.append([f"Día {dia}", producto, unidades, precio_unitario, total_venta, total_con_descuento])

    # Crear un DataFrame con los datos
    columnas = ['Día', 'Producto', 'Unidades', 'Precio Unitario', 'Total Venta', 'Total con Descuento']
    df = pd.DataFrame(datos, columns=columnas)
    return df

# Función para que el usuario ingrese los datos manualmente
def ingresar_ventas():
    productos = []
    print("Ingresa los nombres de los productos (3 productos):")
    for _ in range(3):
        producto = input(f"Nombre del producto {_+1}: ")
        productos.append(producto)

    datos = []
    for dia in range(1, 6):  # Últimos 5 días
        for producto in productos:
            print(f"\nDía {dia} - {producto}")
            unidades = int(input("Ingresa el número de unidades vendidas: "))
            precio_unitario = float(input("Ingresa el precio unitario: "))
            total_venta = unidades * precio_unitario
            descuento = total_venta * 0.05
            total_con_descuento = total_venta - descuento
            datos.append([f"Día {dia}", producto, unidades, precio_unitario, total_venta, total_con_descuento])

    # Crear un DataFrame con los datos
    columnas = ['Día', 'Producto', 'Unidades', 'Precio Unitario', 'Total Venta', 'Total con Descuento']
    df = pd.DataFrame(datos, columns=columnas)
    return df

# Función principal
def main():
    print("¿Cómo deseas obtener los datos de ventas?")
    print("1: Generarlos automáticamente")
    print("2: Ingresarlos manualmente")
    opcion = input("Elige una opción (1 o 2): ")

    if opcion == '1':
        df = generar_ventas()
    elif opcion == '2':
        df = ingresar_ventas()
    else:
        print("Opción no válida. Generando datos automáticamente por defecto.")
        df = generar_ventas()

    # Mostrar el DataFrame resultante
    print("\nDatos de ventas de los últimos 5 días:")
    print(df)

# Ejecutar el programa principal
main()
```

## Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.  El resultado final debe ser una serie con los estadísticos.


```python
# Importar la librería pandas
import pandas as pd

# Función para calcular los estadísticos
def calcular_estadisticos(ingresos):
    """
    Recibe un diccionario con los ingresos mensuales de los clientes.
    Devuelve una serie con el ingreso mínimo, máximo y media.
    """
    # Convertir los valores del diccionario a una serie de pandas
    ingresos_serie = pd.Series(ingresos.values())
    
    # Calcular los estadísticos
    estadisticos = pd.Series({
        'Ingreso Mínimo': ingresos_serie.min(),
        'Ingreso Máximo': ingresos_serie.max(),
        'Ingreso Promedio': ingresos_serie.mean()
    })
    
    return estadisticos

# Ejemplo de uso
ingresos_clientes = {
    'Cliente 1': 1200,
    'Cliente 2': 2500,
    'Cliente 3': 1800,
    'Cliente 4': 3000,
    'Cliente 5': 1700
}

# Llamar a la función y mostrar el resultado
resultado = calcular_estadisticos(ingresos_clientes)
print("Estadísticos de los ingresos:")
print(resultado)
```

## Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla: 
## Mes Ventas Gastos 
## Enero 30500 22000 
## Febrero 35600 23450 
## Marzo 28300 18100 
## Abril 33900 35700 
## Mayo 42500 32450 
## A continuación, generar nuevas columna: 
## 1. Columna 1 = Gasto/Ventas*100. 
## 2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”. 
## 3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto. 4. Obtener el total de las ventas de los últimos 3 meses 5. Obtener el total de gastos de febrero a abril


```python
# Importar la librería pandas
import pandas as pd

# Crear el DataFrame inicial
datos = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo"],
    "Ventas": [30500, 35600, 28300, 33900, 42500],
    "Gastos": [22000, 23450, 18100, 35700, 32450]
}
df = pd.DataFrame(datos)

# 1. Generar la columna Gasto/Ventas*100
df["Gasto/Ventas (%)"] = (df["Gastos"] / df["Ventas"]) * 100

# 2. Generar la columna para verificar si la meta fue superada
df["Meta Superada"] = df["Ventas"].apply(lambda x: "Meta Superada" if x >= 30000 else "Meta no Superada")

# 3. Generar la columna para calcular el sobre gasto
df["Sobre Gasto"] = df.apply(lambda row: row["Gastos"] - row["Ventas"] if row["Gastos"] > row["Ventas"] else 0, axis=1)

# 4. Calcular el total de ventas de los últimos 3 meses
total_ventas_ultimos_3_meses = df["Ventas"][-3:].sum()

# 5. Calcular el total de gastos de febrero a abril
gastos_febrero_abril = df.loc[1:3, "Gastos"].sum()

# Mostrar el DataFrame modificado y los resultados adicionales
print("DataFrame modificado:")
print(df)

print("\nTotal de ventas de los últimos 3 meses:", total_ventas_ultimos_3_meses)
print("Total de gastos de febrero a abril:", gastos_febrero_abril)
```

## Con el siguiente fichero datos_seguros.csv contiene información de riesgo de clientes de un banco. Escribir un programa con los siguientes requisitos:  1. Extraer las 1000 primeras filas y almacenarlas en un contenedor.  4. Tabla de doble entrada y media. 


```python
import pandas as pd

# Cargar el archivo CSV con el delimitador correcto
ruta_archivo = 'datos_seguros.csv'
df = pd.read_csv(ruta_archivo, sep=";")

# Verificar la estructura del DataFrame
print("Dimensiones del DataFrame:", df.shape)
print("Número total de datos:", df.size)
print("Nombres de columnas:", df.columns.tolist())
print("Primeros 10 índices (filas):", df.index[:10].tolist())
print("\nTipos de datos de las columnas:\n", df.dtypes)

# Mostrar las 10 primeras y 10 últimas filas
print("\nPrimeras 10 filas:")
print(df.head(10))
print("\nÚltimas 10 filas:")
print(df.tail(10))

# Verificar si hay datos nulos
print("\n¿Datos nulos en el DataFrame?")
print(df.isnull().sum())

# Extraer las primeras 1000 filas
contenedor = df.head(1000)
print("\nContenedor de las primeras 1000 filas:")
print(contenedor.info())

# Crear la tabla de doble entrada y calcular la media
# Reemplazar 'genero', 'departamento', 'saldo_bco_t2' con los nombres correctos de las columnas
columna1 = 'SEXO'  # Cambiar si tiene otro nombre
columna2 = 'DEPARTAMENTO'  # Cambiar si tiene otro nombre
columna_numerica = 'SALDO_BCO_T2'  # Cambiar si tiene otro nombre

tabla_doble_entrada = pd.crosstab(df[columna1], df[columna2], values=df[columna_numerica], aggfunc='mean')

# Mostrar la tabla de doble entrada
print("\nTabla de doble entrada con la media:")
print(tabla_doble_entrada)
```
